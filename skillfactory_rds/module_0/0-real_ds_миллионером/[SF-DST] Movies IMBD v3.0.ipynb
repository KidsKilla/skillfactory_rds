```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# import itertools
import os
from collections import Counter
print(os.listdir("./"))
```

```python
data = pd.read_csv('./data.csv')
data.head(5)
```

```python
def split(str):
    return str.strip().split('\n');
def select (ids, col='imdb_id'):
    return data[data[col].isin(ids)]
def get_ids (str):
    return list(map(lambda line: line[-9::], split(str)))
def add_profit (df):
    df.profit = (df.revenue - df.budget)
    return df
norm_c_name = [
    'warner',
    'universal',
    'paramount',
    'columbia',
    'fox',
    'disney',
    'lucasfilm',
    'jim henson',
    'four by two'
]
def norm_company_name(name):
    n = name.lower()
    comps = list(filter(lambda c_name: c_name in n, norm_c_name))
    if len(comps) > 0:
        return comps[0]
    return name

def norm_comps(comps):
    return list(map(norm_company_name, comps))

def get_data(df = data):
    df = df.copy()
    add_profit(df)
    df.genres = df.genres.str.split('|')
    df.cast = df.cast.str.split('|')
    df.production_companies = df.production_companies.str.split('|').apply(norm_comps)
    return df

def flatten(s):
    return s.apply(pd.Series).stack().reset_index(drop=True)

print(get_data())
print(data.info())
print(data.describe())
print(len(data))
```
